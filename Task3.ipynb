```python
# Input dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Step 1: Find the highest performance score
highest_score = max(employees.values())
# Step 2: Find all employees who have the highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
# Step 3: Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
import string
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))
# Split into words
words = query.split()
# Count frequency
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Filter words searched more than once
result = {word: count for word, count in word_count.items() if count > 1}
print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
# Loop through readings with index
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even (valid)
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
# Extract domain and count
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total_users = len(emails)
# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3  # 30% above average
# Detect spikes
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = {}
# Count occurrences
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1
# Display duplicates only
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
